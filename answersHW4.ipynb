```python
#Question 1
import matplotlib.pyplot as plt

figure, ax = plt.subplots()
ax.scatter(env['ELEVATION'], env['COVER_ROCK'])
slope, intercept = np.polyfit(env['ELEVATION'], env['COVER_ROCK'], 1)
x = np.array([env['ELEVATION'].min(), env['ELEVATION'].max()])
y = slope*x + intercept
ax.plot(x,y, color = 'Red')
ax.set_ylabel("Rock Cover %")
ax.set_xlabel("Elevation (m)")
ax.set_title('The Relationship between Elevation and Rock Cover %')
plt.savefig('/Users/liamzeeum/Github')

#There appears to be no clear relationship between elevation and rock cover



```
